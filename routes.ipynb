# Partie 2 : Analyse

## Quels sont les 5 natures de travaux les plus fréquentes?

```python
dataframe_nature=route1.groupby('naturedernierstravaux').entityid.count().sort_values(ascending=False)
```

```python
dataframe_nature = dataframe_nature.head(5)
dataframe_nature
```

```python
import numpy as np
import matplotlib.pyplot as plt
ax = dataframe_nature.plot(kind = "bar", figsize=(10,10))
ax.set_xlabel("nature derniers travaux", fontsize=16)
ax.set_title("Count", fontsize=16)
plt.ylim([0, 1500])
plt.grid(False)
plt.show()
ax
```

## Pour chaque année, quel est le nombre de travaux réalisés et l’épaisseur moyenne? Triez le résultat par épaisseur descendante.


```python
dataframe_annee=route1.groupby('anneedernierstravaux').agg({'epaisseurdernierstravaux': [np.size, np.mean]})
```

```python
df = dataframe_annee.sort_values([('epaisseurdernierstravaux', 'size')], ascending=False)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
ax = df.plot(kind = "bar", figsize=(10,10))
ax.set_xlabel("annee", fontsize=20)
ax.set_title("nombre de travaux et moyenne epaisseur", fontsize=16)
plt.ylim([0, 500])
plt.grid(False)
plt.show()
ax
```
